```python
# positional argument
def calculation (a , b):
  print("addition" , a+b)
  print("subtraction" , a-b)
  print("multiplication" , a*b)
  print("Division" , a/b)
  return "calculation is complete"
```

```python
print(calculation(10 , 5))
```
