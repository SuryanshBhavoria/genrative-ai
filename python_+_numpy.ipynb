```python
# default argument
def area_perimeter(width=4 , height=8):
  area = width*height
  perimeter = 2*(width+height)
  return "Area is " , area , "perimeter is" , perimeter
```

```python
print(area_perimeter())   # if we does not give argument so by default it take value from above.
```

```python
# variable length argument
# non keyword positional variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2 , 4, 3, 5, 6 , 3 , 2 ,5)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num #sum = sum + num
  return sum
```

```python
sum_number(3 , 4 , 2 , 42, 2 , 5 , 8)
```

```python
#keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(type(kwargs))
  print(len(kwargs))
```

```python
def pass_student(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_student(A = 30 , B = 40 , C = 50 , D = 80 , E = 20)
```
